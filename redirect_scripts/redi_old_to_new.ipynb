```python
import pickle
import networkx as nx
```

```python
# open the pickle file as a graph
redi_graph = pickle.load(open("cc_sample_2_redirect_graph.p", "rb"), encoding='utf-8')
redi_graph
```

```python
# find all connected components in a directed graph
# that are greater than 1 node
# and save them as a list of lists
redi_components = [c for c in nx.weakly_connected_components(redi_graph) if len(c) > 1]
len(redi_components)
```

```python
# all nodes that have no ingoing or outgoing edges
leaf_nodes = [n for n in redi_graph.nodes if redi_graph.in_degree(n) == 0 and redi_graph.out_degree(n) == 0]
len(leaf_nodes)
```

```python
import csv
with open("tesssst.csv", 'w') as csvfile:
    writer = csv.writer(csvfile)
    for l in leaf_nodes:
        writer.writerow([l])
```

```python
count_redirect_until_timeout = 0
count_redirect_until_not_found = 0
count_redirect_until_error = 0
count_redirect_until_found = 0
```

```python
for c in redi_components:
    c = list(c)
    remarks = []
    for node in c:
        remarks.append(redi_graph.nodes[node]['remark'])
    if 'redirect_until_timeout' in remarks or 'timeout' in remarks:
        count_redirect_until_timeout += 1
    if 'redirect_until_not_found' in remarks or 'not_found' in remarks:
        count_redirect_until_not_found += 1
    if 'redirect_until_error' in remarks or 'error' in remarks:
        count_redirect_until_error += 1
    if 'redirect_until_found' in remarks or 'found_without_redirect' in remarks:
        count_redirect_until_found += 1
```

```python
display(count_redirect_until_timeout, count_redirect_until_not_found, count_redirect_until_error, count_redirect_until_found)
```
